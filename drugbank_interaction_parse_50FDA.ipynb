```python
import xml.etree.ElementTree as ET 
tree = ET.ElementTree(file="full_database.xml")
```

```python
root = tree.getroot() 
```

```python
druglist=['Ifosfamide','Amifostine', 'Chlorambucil',
 'Hydroxyurea', 'Estramustine phosphate sodium',
 'Exemestane', 'Thiotepa', 'Thalidomide',
 'Thioguanine', 'Gefitinib' ,'Sorafenib tosylate',
 'Procarbazine hydrochloride', 'Melphalan', 'Ruxolitinib',
 'Teniposide', 'Oxaliplatin', 'Pemetrexed Disodium',
 'Fulvestrant', 'Vinorelbine tartrate' ,'Crizotinib',
 'Axitinib', 'Valrubicin', 'Dactinomycin',
 'Romidepsin', 'Carmustine' ,'Mitotane',
 'Erlotinib hydrochloride', 'Vandetanib', 'Cladribine',
 'Daunorubicin hydrochloride', 'Vismodegib',
 'Megestrol acetate', 'Lenalidomide', 'Tamoxifen citrate',
 'Quinacrine hydrochloride' ,'Bortezomib',
 'Vincristine sulfate', 'Uracil mustard', 'Dacarbazine',
 'Lomustine', 'Mitoxantrone', 'Cytarabine hydrochloride',
 'Arsenic trioxide' ,'Vinblastine sulfate', 'Cabazitaxel',
 '2-Fluoro Ara-A', 'Methoxsalen' ,'Everolimus',
 'pralatrexate' ,'Floxuridine']
```

```python
pairList=[]
for x in root.findall('{http://www.drugbank.ca}drug'):
    name =x.find('{http://www.drugbank.ca}name').text
    if name in druglist:
        print(x,name)
        print("========================================================================")
        for elem in x.iter(tag='{http://www.drugbank.ca}drug-interaction'):
            #for child in elem:
            if elem.find("{http://www.drugbank.ca}name").text in druglist:
                print(elem.find("{http://www.drugbank.ca}name").text)
                pairList.append(set((name,elem.find("{http://www.drugbank.ca}name").text)))
```

```python
# del repeated interactions
temp = []
for item in pairList:
    if not item in temp:
        temp.append(item)
print(len(temp))
```

```python
import pickle
fw = open('drugbank_interactions_50FDA','wb')
pickle.dump(temp,fw)
fw.close()
```
